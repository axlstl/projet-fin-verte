```python
portfolio_social = pd.DataFrame(columns=["Date", "Portfolio_Data"])
portfolio_environnement = pd.DataFrame(columns=["Date", "Portfolio_Data"])
portfolio_ESG = pd.DataFrame(columns=["Date", "Portfolio_Data"])
```

```python
PB = PB(df_esg)
```

```python
portefeuille_environnement = PB.environnement()
portefeuille_social = PB.social()
portefeuille_ESG = PB.ESG()
```

```python
df_esg["ratingDate"] = pd.to_datetime(df_esg["ratingYear"].astype(str) + "-" + df_esg["ratingMonth"].astype(str) + "-01")
most_recent_date = df_esg["ratingDate"].max()

# Liste des portefeuilles
portfolios = {
    "portfolio_environnement": portfolio_environnement,
    "portfolio_social": portfolio_social,
    "portfolio_ESG": portfolio_ESG
}

# Mise Ã  jour les portefeuilles
for name, portfolio in portfolios.items():
    if portfolio.empty or most_recent_date.strftime('%Y-%m') != portfolio["Date"].max():
        new_row = pd.DataFrame([{
            "Date": most_recent_date.strftime('%Y-%m'),
            "Portfolio_Data": portefeuille_environnement if name == "portfolio_environnement" else
                             portefeuille_social if name == "portfolio_social" else
                             portefeuille_ESG
        }])
        portfolios[name] = pd.concat([portfolio, new_row], ignore_index=True)

portfolio_environnement = portfolios["portfolio_environnement"]
portfolio_social = portfolios["portfolio_social"]
portfolio_ESG = portfolios["portfolio_ESG"]
```

```python
# CrÃ©ation d'une liste pour stocker les portefeuilles au bon format
portfolios_to_analyze = []

# Pour chaque portefeuille
for portfolio_df, name in zip([portfolio_environnement, portfolio_social, portfolio_ESG], 
                            ["Environnement", "Social", "ESG"]):
    if not portfolio_df.empty:
        # On prend le portefeuille le plus rÃ©cent
        last_entry = portfolio_df.iloc[-1]
        
        # Construction du dictionnaire au format attendu
        portfolio_dict = {
            "name": name,
            "tickers": list(last_entry["Portfolio_Data"].keys()),
            "weights": last_entry["Portfolio_Data"],
            "investment_start_date": last_entry["Date"] + "-01"  # Convertir YYYY-MM en YYYY-MM-DD
        }
        portfolios_to_analyze.append(portfolio_dict)
```

```python
# Analyser les portefeuilles avec le bon format
analyzer = PortfolioPerformanceAnalyzer()
resultats = analyzer.analyze_portfolios(portfolios_to_analyze)
```
